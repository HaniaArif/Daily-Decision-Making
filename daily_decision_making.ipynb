Think of a decision you make every day or often. Write code that demonstrates
the logic of decisions of the making process using Python. 

Use: 2 inputs If, elif, else Comparison operators Random F strings List Loops

```python
import random

#list of clothes according to weather
#shirts
light_shirts = ["plain white shirt", 
                "plain blue shirt", 
                "plain white T-shirt", 
                "white T-Shirt with Avengers logo"]

moderate_shirts = ["blue-black Addidas T-shirt", 
                   "plain brown T-shirt", 
                   "black and white check Shirt"]

warm_shirts = ["black hoodie", 
               "white hoodie with Dragon logo", 
               "dark green double shirt"]
#pants
all_pants = ["blue Jeans", 
         "white trousers", 
         "black Jeans"]

#Jackets
all_jackets = ["blue denim jacket", 
           "white and black jacket with pockets"]

#-------STEP 1: select today's weather-------
weather = -1 #not selected yet
while(True):
  weather = int(input("What is the weather today?\n 1. Sunny\n 2. Cloudy\n 3. Rainy\n 4. Snow falling\n Select weather: "))
  if(weather >=1 and weather <=4):
    break
  else:
    print("Please select one of the given options")


selected_shirt = None   #not selected yet
selected_pants = None   #not selected yet
selected_jacket = None  #not selected yet


#-------STEP 2: show available shirts and select a random shirt from the available ones-------
while(True):
  print("\n\nYou have these shirts to wear:")
  if(weather == 1):   #sunny
    i = 1
    for shirt in light_shirts:
      print(f"{i}. {shirt}")
      i = i+1
    selected_shirt = random.choice(light_shirts)
    
  elif(weather == 2): #cloudy
    i = 1
    for shirt in moderate_shirts:
      print(f"{i}. {shirt}")
      i = i+1
    selected_shirt = random.choice(moderate_shirts)

  elif(weather == 3): #rainy
    i = 1
    for shirt in moderate_shirts:
      print(f"{i}. {shirt}")
      i = i+1
    selected_shirt = random.choice(moderate_shirts)

  else:               #snow falling
    i = 1
    for shirt in warm_shirts:
      print(f"{i}. {shirt}")
      i = i+1
    selected_shirt = random.choice(warm_shirts)
  
  print("Selected shirt: ", selected_shirt)
  decision = None     #not decided yet
  print("Decide: \n 1. Wear this\n 2. Select again")
  decision = int(input("Your Decision: "))
  if(decision == 1):  #if decided to wear this shirt then move to the next selection
    break
  #if decided to selected again then go through the loop again


#-------STEP 3: show available pants and select a random pant from the available ones-------
while(True):
  print("\n\nYou have these pants to wear:")
  i = 1
  for pants in all_pants:
    print(f"{i}. {pants}")
  selected_pants = random.choice(all_pants)
  print("Selected pants: ", selected_pants)
  decision = None     #not decided yet
  print("Decide: \n 1. Wear this\n 2. Select again")
  decision = int(input("Your Decision: "))
  if(decision == 1):  #if decided to wear this pants then move to the next selection
    break
  #if decided to selected again then go through the loop again


#-------STEP 4: show available jackets and select a random jacket from the available ones-------
while(True):
  print("\n\nYou have these jackets to wear:")
  i = 1
  for jacket in all_jackets:
    print(f"{i}. {jacket}")
  selected_jacket = random.choice(all_jackets)
  print("Selected pants: ", selected_jacket)
  decision = None     #not decided yet
  print("Decide: \n 1. Wear this\n 2. Select again")
  decision = int(input("Your Decision: "))
  if(decision == 1):  #if decided to wear this jacket then move to the next selection
    break
  #if decided to selected again then go through the loop again

#-------STEP 5: decode selected weather number for printing-------
selected_weather = None
if(weather == 1):
  selected_weather = "Sunny"

elif(weather == 2):
  selected_weather = "Cloudy"

elif(weather == 3):
  selected_weather = "Rainy"

else:
  selected_weather = "Snow Falling"

#-------STEP 6: show the selected clothes that you have to wear today according to the weather-------
print(f"\n\nIt was {selected_weather} today and you are goint to wear:\n Shirt: {selected_shirt}\n Pants: {selected_pants}\n Jacket: {selected_jacket}")
```
